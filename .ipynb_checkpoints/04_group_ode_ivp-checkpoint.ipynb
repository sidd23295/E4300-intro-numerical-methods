## Question 1 - Warm Up

Consider the initial value problem
$$
    x''(x) = \cos(xt), ~~~ x(0) = 1, ~~~ x'(0) = 0, ~~ 0\leq t \leq 1
$$

**(a)** Reformulate the problem as an IVP for a system of ODEs

$$\vec{u} = \begin{bmatrix} x \\ x' \end{bmatrix} = \begin{bmatrix} u_1 \\ u_2 \end{bmatrix}$$

$$\frac{\text{d}}{\text{d}t} \begin{bmatrix} u_1 \\ u_2 \end{bmatrix} = \begin{bmatrix} u_2 \\ cos(u_1 t) \end{bmatrix} $$

**(b)** Plot the trajectory using Euler's method, RK2, and RK4.

```python
import scipy.integrate as integrate

def f(t, u):
    return numpy.array([u[1], numpy.cos(u[0]*t)])

# Implement Forward Euler
t = numpy.linspace(0.0, 5, 100)
delta_t = t[1] - t[0]
u_euler = numpy.empty((2, 100))
u_euler[0,:] = 1
u_euler[1,:] = 0
for (n, t_n) in enumerate(t[:-1]):
    u_euler[:,n + 1] = u_euler[:,n] + delta_t * f(t_n, u_euler[:,n])

# Implement RK-2 & RK-4
u_2 = numpy.empty((2, 100))
u_2[0,0] = 1.0
u_2[1,0] = 0.0
u_4 = numpy.empty((2, 100))
u_4[0,0] = 1.0
u_4[1,0] = 0.0
for (n, t_n) in enumerate(t[1:]):
    u_2[:,n+1] = u_2[:,n] + 0.5 * delta_t * f(t_n, u_2[:,n])
    u_2[:,n+1] = u_2[:,n] + delta_t * f(t_n, u_2[:,n+1])
    y_1 = u_4[:,n]
    y_2 = u_4[:,n] + 0.5 * delta_t * f(t_n + 0.5 * delta_t, y_1)
    y_3 = u_4[:,n] + 0.5 * delta_t * f(t_n + 0.5, y_2)
    y_4 = u_4[:,n] + delta_t * f(t_n + 0.5 * delta_t, y_3)
    u_4[:,n+1] = u_4[:,n] + delta_t / 6.0 * (f(t_n, y_1) + 2.0 * f(t_n + 0.5 * delta_t, y_2) + 2.0 * f(t_n + 0.5 * delta_t, y_3) + f(t_n + delta_t, y_4))

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
#fig.set_figwidth(fig.get_figwidth() + 1.1)

axes.plot(t, u_euler[0,:],'b', label='Forward Euler')
axes.plot(t,u_euler[1,:],'b--', label='Forward Euler')
axes.plot(t, u_2[0,:],'r', t,u_2[1,:],'r--', label='2-stage')
axes.plot(t, u_4[0,:],'k', t,u_4[1,:],'k--', label='4-stage')
axes.set_title("Solution")
axes.set_xlabel("t")
axes.set_ylabel("y(t)")
axes.legend(loc='center left',bbox_to_anchor=(1,0.5))


fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
#fig.set_figwidth(fig.get_figwidth() + 1.1)

axes.plot(u_euler[0,:], u_euler[1,:],'b',label='Forward Euler')
axes.plot(u_2[0,:], u_2[1,:],'r',label='2-stage')
axes.plot(u_4[0,:], u_4[1,:],'k',label='4-stage')
axes.set_title("Phase Diagram")
axes.set_xlabel("y(t)")
axes.set_ylabel("y'(t)")
axes.legend(loc='center left',bbox_to_anchor=(1,0.5))

plt.show()
```

## Question 2 - Van der Pol Oscillator

Lets find a solution to another oscillator IVP ODE problem just like in the notes (see solution to Van der Pol Oscillator). It is called a Rayleigh Oscillator and is closely related to the Rayleigh differential equation.  The main difference is that the $x^2$ term  is replaced with a $\dot{x}^2$ term.  

Each system has the property that the origin is an unstable equilibrium point and trajectories near it spiral outward until either the $x^2$ or $\dot{x}^2$ get too large and are offset by damping.  

Here is an elegant version of a forced Rayleigh Oscillator with a chaotic solution:
$$ 
    \ddot{x} + (\dot{x}^2-4)\dot{x} + x = 5 \sin(4t)
$$
with the initial condition $(x_0, v_0, t_0) = (-0.9, 4, 0)$.

For more details see Sprott, Julien Clinton. "Elegant Chaos: Algebraically Simple Chaotic Flows." World Scientific

**(a)** Reformulate this problem as an IVP for a system of ODEs. 

$$\vec{u} = \begin{bmatrix} x \\ x' \end{bmatrix} = \begin{bmatrix} u_1 \\ u_2 \end{bmatrix}$$

$$\frac{\text{d}}{\text{d}t} \begin{bmatrix} u_1 \\ u_2 \end{bmatrix} = \begin{bmatrix} u_2 \\ 5 sin(4t) - (u_2^2-4)u_2 - u_1 \end{bmatrix} $$

**(b)** Solve this problem now using a 4 Step Adams-Bashforth using RK2 to start the method.

```python
import scipy.integrate as integrate

def f(t, u):
    return numpy.array([u[1], 5.0*numpy.sin(4.0*t)-(u[1]**2-4.0)*u[1]-u[0]])

t = numpy.linspace(0.0, 10, 300)
delta_t = t[1] - t[0]

u_4 = numpy.empty((2, 300))
u_4[0,0] = -0.9
u_4[1,0] = 4.0
for (n, t_n) in enumerate(t[1:]):
    y_1 = u_4[:,n]
    y_2 = u_4[:,n] + 0.5 * delta_t * f(t_n + 0.5 * delta_t, y_1)
    y_3 = u_4[:,n] + 0.5 * delta_t * f(t_n + 0.5, y_2)
    y_4 = u_4[:,n] + delta_t * f(t_n + 0.5 * delta_t, y_3)
    u_4[:,n+1] = u_4[:,n] + delta_t / 6.0 * (f(t_n, y_1) + 2.0 * f(t_n + 0.5 * delta_t, y_2) + 2.0 * f(t_n + 0.5 * delta_t, y_3) + f(t_n + delta_t, y_4))


u_ab4 = numpy.empty((2, 300))

# Use Rk-2 to start the method
# And then use AB-4
u_ab4[0,0] = -0.9
u_ab4[1,0] = 4.0
u_ab4[:,1] = u_ab4[:,0] + 0.5 * delta_t * f(t[0], u_ab4[:,0])
u_ab4[:,1] = u_ab4[:,0] + delta_t * f(t[0], u_ab4[:,1])
u_ab4[:,2] = u_ab4[:,1] + 0.5 * delta_t * f(t[0], u_ab4[:,1])
u_ab4[:,2] = u_ab4[:,1] + delta_t * f(t[0], u_ab4[:,2])
u_ab4[:,3] = u_ab4[:,2] + 0.5 * delta_t * f(t[0], u_ab4[:,2])
u_ab4[:,3] = u_ab4[:,2] + delta_t * f(t[0], u_ab4[:,3])
for n in xrange(0,len(t)-4):
    u_ab4[:,n+4] = u_ab4[:,n + 3] + delta_t / 24.0 * (-9.0* f(t[n], u_ab4[:,n]) + 37.0 * f(t[n+1], u_ab4[:,n+1]) - 59.0 * f(t[n+2], u_ab4[:,n+2]) + 55.0 * f(t[n+3], u_ab4[:,n+3]))
    

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.plot(u_4[0,:], u_4[1,:],'r',label='RK4-stage')
axes.set_title("Phase Diagram")
axes.set_xlabel("y(t)")
axes.set_ylabel("y'(t)")
axes.legend(loc='center left',bbox_to_anchor=(1,0.5))

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.plot(u_ab4[0,:], u_ab4[1,:],'r',label='AB4-stage')
axes.set_title("Phase Diagram")
axes.set_xlabel("y(t)")
axes.set_ylabel("y'(t)")
axes.legend(loc='center left',bbox_to_anchor=(1,0.5))

plt.show()
```

## Question 3 - Absolute Stability Regions and Order Stars

**(a)** Plot the regions of absolute stability for the following:

For the Taylor series methods also determine the interior and exterior of the stability regions.

```python
# plotting code
fig = plt.figure(figsize=(8, 8))
fig.set_figwidth(fig.get_figwidth() * 2.0)
fig.set_figheight(fig.get_figheight() * 2.0)

#============================================================================
# Taylor Series method of order 2
axes = fig.add_subplot(4, 2, 1)
x = numpy.linspace(-5, 5, 100)
y = numpy.linspace(-5, 5, 100)
X, Y = numpy.meshgrid(x, y)
Z = X + 1j * Y

# Evaluate which regions are actually in stability region
print "Taylor series method of order 2 - Stability region tests"
z = -1 + 1j * 0
print "  Inside of strange region: ", numpy.abs(1.0 + z)
z = -4 + 1j * 0
print "  Outside of strange region: ", numpy.abs(1.0 + z)

axes.contour(X, Y, numpy.abs(1 + Z + 0.5 * Z**2), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_aspect('equal')
axes.set_title("2nd Order Taylor Series")
#============================================================================


#============================================================================
# Taylor Series method of order 3
axes = fig.add_subplot(4, 2, 2)
x = numpy.linspace(-5, 5, 100)
y = numpy.linspace(-5, 5, 100)
X, Y = numpy.meshgrid(x, y)
Z = X + 1j * Y

print "Taylor series method of order 3 - Stability region tests"
z = -1 + 1j * 0
print "  Inside of strange region: ", numpy.abs(1.0 + z)
z = -4 + 1j * 0
print "  Outside of strange region: ", numpy.abs(1.0 + z)

axes.contour(X, Y, numpy.abs(1 + Z + 0.5 * Z**2 + 1.0/6.0 * Z**3), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_aspect('equal')
axes.set_title("3rd Order Taylor Series")
#============================================================================


#============================================================================
# Taylor Series method of order 4
axes = fig.add_subplot(4, 2, 3)
x = numpy.linspace(-5, 5, 100)
y = numpy.linspace(-5, 5, 100)
X, Y = numpy.meshgrid(x, y)
Z = X + 1j * Y

print "Taylor series method of order 4 - Stability region tests"
z = -1 + 1j * 0
print "  Inside of strange region: ", numpy.abs(1.0 + z)
z = -4 + 1j * 0
print "  Outside of strange region: ", numpy.abs(1.0 + z)

axes.contour(X, Y, numpy.abs(1 + Z + 0.5 * Z**2 + 1.0/6.0 * Z**3 + 1.0 / 24.0 * Z**4), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_aspect('equal')
axes.set_title("4th Order Taylor Series")
#============================================================================


#============================================================================
# Taylor Series method of order 5
axes = fig.add_subplot(4, 2, 4)
x = numpy.linspace(-5, 5, 100)
y = numpy.linspace(-5, 5, 100)
X, Y = numpy.meshgrid(x, y)
Z = X + 1j * Y

print "Taylor series method of order 5 - Stability region tests"
z = -1 + 1j * 0
print "  Inside of strange region: ", numpy.abs(1.0 + z)
z = -4 + 1j * 0
print "  Outside of strange region: ", numpy.abs(1.0 + z)

axes.contour(X, Y, numpy.abs(1 + Z + 0.5 * Z**2 + 1.0/6.0 * Z**3 + 1.0 / 24.0 * Z**4 + 1.0 / 120.0 * Z**5), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_aspect('equal')
axes.set_title("5th Order Taylor Series")
#============================================================================


#============================================================================
# 3-step Adams-Bashforth
theta = numpy.linspace(0.0, 2.0 * numpy.pi, 1000)
xi = numpy.exp(1j * theta)

rho_3AB = lambda xi: (xi - 1.0) * xi**2
sigma_3AB = lambda xi: (5.0 - 16.0 * xi + 23.0 * xi**2 ) / 12.0
z_3AB = rho_3AB(xi) / sigma_3AB(xi)

axes = fig.add_subplot(4, 2, 5)
axes.plot(z_3AB.real, z_3AB.imag)
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_title("3-step Adams-Bashforth")
axes.set_aspect('equal')
axes.set_xlim([-2, 3])
axes.set_ylim([-2, 2])
#============================================================================


#============================================================================
# 4-step Adams-Bashforth
theta = numpy.linspace(0.0, 2.0 * numpy.pi, 1000)
xi = numpy.exp(1j * theta)

rho_4AB = lambda xi: (xi - 1.0) * xi**3
sigma_4AB = lambda xi: (-9.0 + 37.0 * xi - 59.0 * xi**2 + 55.0 * xi**3 ) / 24.0
z_4AB = rho_4AB(xi) / sigma_4AB(xi)

axes = fig.add_subplot(4, 2, 6)
axes.plot(z_4AB.real, z_4AB.imag)
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_title("4-step Adams-Bashforth")
axes.set_aspect('equal')
axes.set_xlim([-2, 3])
axes.set_ylim([-2, 2])
#============================================================================


#============================================================================
# 3-step Adams-Moulton
theta = numpy.linspace(0.0, 2.0 * numpy.pi, 1000)
xi = numpy.exp(1j * theta)

rho_3AM = lambda xi: (xi - 1.0) * xi**2
sigma_3AM = lambda xi: (1.0 - 5.0 * xi + 19.0 * xi**2 + 9.0 * xi**3 ) / 24.0
z_3AM = rho_3AM(xi) / sigma_3AM(xi)

axes = fig.add_subplot(4, 2, 7)
axes.plot(z_3AM.real, z_3AM.imag)
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_title("3-step Adams-Moulton")
axes.set_aspect('equal')
axes.set_xlim([-3.5, 3])
axes.set_ylim([-2, 2])
#============================================================================


#============================================================================
# 4-step Adams-Moulton
theta = numpy.linspace(0.0, 2.0 * numpy.pi, 1000)
xi = numpy.exp(1j * theta)

rho_4AM = lambda xi: (xi - 1.0) * xi**3
sigma_4AM = lambda xi: (-19.0 + 106.0 * xi - 264.0 * xi**2 + 646.0 * xi**3 + 251.0 * xi**4 ) / 720.0
z_4AM = rho_4AM(xi) / sigma_4AM(xi)

axes = fig.add_subplot(4, 2, 8)
axes.plot(z_4AM.real, z_4AM.imag)
axes.plot(x, numpy.zeros(x.shape),'k')
axes.plot(numpy.zeros(y.shape), y,'k')
axes.set_title("4-step Adams-Moulton")
axes.set_aspect('equal')
axes.set_xlim([-2, 3])
axes.set_ylim([-2, 2])
#============================================================================



plt.show()
```

**(b)** The region of absolute stability as defined ensures that the error does not grow in subsequent time-steps hence why we require $|R(z)| < 1$.  In reality what we really want is that the errors decay faster than the solution or that it grows slower than the true solution.  Consider the solution of an ODE $u(t) = e^{\lambda t}$, if $\lambda < 0$ then we want our errors made at each time step $E^n$ to decay faster than $e^{\lambda t}$ so that $|E^n| < e^{\lambda t}$.  Conversely if $\lambda > 0$ then we might also want $|E^n| < e^{\lambda t}$.  This suggests a new definition of stability, called the **relative stability** defined as

$$|R(z)| \leq |e^z|.$$

These turn out to be difficult to plot but proved pivotal due to work in 1978 by Wanner, Hairer and Nrset.  More recently a new set of regions, called **order stars**, have become more popular to consider.  These are defined as the three regions

$$\begin{aligned}
    \mathcal{A}_{-} &= \{ z \in \mathbb{C}: |R(z)| < |e^z| \} = \{z \in \mathbb{C}: |e^{-z} R(z)| < 1\}, \\ 
    \mathcal{A}_{0} &= \{ z \in \mathbb{C}: |R(z)| = |e^z| \} = \{z \in \mathbb{C}: |e^{-z} R(z)| = 1\}, ~~~~\text{and} \\
    \mathcal{A}_{+} &= \{ z \in \mathbb{C}: |R(z)| > |e^z| \} = \{z \in \mathbb{C}: |e^{-z} R(z)| > 1\}.
\end{aligned}$$

Plot the region $\mathcal{A}_{0}$ for the Taylor series methods asked for in part (a).  Do you see any relation to the number of "fingers" and the order of the method?

```python
x = numpy.linspace(-5,5,100)
y = numpy.linspace(-5,5,100)
X, Y = numpy.meshgrid(x,y)
Z = X + 1j * Y

#Taylor Series Methods for orders of 2 3 4 and 5. 
R_2 = lambda z: 1.0 + z + 0.5 * z**2
R_3 = lambda z: 1.0 + z + 0.5 * z**2 + 1.0 / 6.0 * z**3
R_4 = lambda z: 1.0 + z + 0.5 * z**2 + 1.0 / 6.0 * z**3 + 1.0 / 24.0 * z**4
R_5 = lambda z: 1.0 + z + 0.5 * z**2 + 1.0 / 6.0 * z**3 + 1.0 / 24.0 * z**4 + 1.0 / 120.0 * z**5


#plotting code
fig = plt.figure(figsize=(8, 8))
fig.set_figwidth(fig.get_figwidth() * 2.0)

axes = fig.add_subplot(2, 2, 1)
axes.contour(X,Y, numpy.abs(numpy.exp(-Z) * R_2(Z)), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape), 'k')
axes.plot(numpy.zeros(y.shape), y, 'k')
axes.set_title("Taylor Series Method of Order 2")
axes.set_xlabel("x")
axes.set_ylabel("y")

axes = fig.add_subplot(2, 2, 2)
axes.contour(X,Y, numpy.abs(numpy.exp(-Z) * R_3(Z)), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape), 'k')
axes.plot(numpy.zeros(y.shape), y, 'k')
axes.set_title("Taylor Series Method of Order 3")
axes.set_xlabel("x")
axes.set_ylabel("y")

axes = fig.add_subplot(2, 2, 3)
axes.contour(X,Y, numpy.abs(numpy.exp(-Z) * R_4(Z)), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape), 'k')
axes.plot(numpy.zeros(y.shape), y, 'k')
axes.set_title("Taylor Series Method of Order 4")
axes.set_xlabel("x")
axes.set_ylabel("y")

axes = fig.add_subplot(2, 2, 4)
axes.contour(X,Y, numpy.abs(numpy.exp(-Z) * R_5(Z)), levels=[1.0])
axes.plot(x, numpy.zeros(x.shape), 'k')
axes.plot(numpy.zeros(y.shape), y, 'k')
axes.set_title("Taylor Series Method of Order 5")
axes.set_xlabel("x")
axes.set_ylabel("y")

print "The number of fingers are same as the order-number of the method."

plt.show()
```
